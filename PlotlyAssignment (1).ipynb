```python
## Answer 1)
import seaborn as sns
import plotly.express as px

# Load the "titanic" dataset
titanic = sns.load_dataset("titanic")

# Plot scatter plot using Plotly Express
fig = px.scatter(titanic, x='age', y='fare')

# Update plot layout and display the figure
fig.update_layout(title="Scatter Plot of Age vs Fare",
                  xaxis_title="Age",
                  yaxis_title="Fare")

fig.show()
```

```python
## Answer 2)
import plotly.express as px
import plotly.io as pio

# Load the "tips" dataset from Plotly
tips = px.data.tips()

# Plot box plot using Plotly Express
fig = px.box(tips, x='day', y='total_bill')

# Update plot layout and display the figure
fig.update_layout(title="Box Plot of Total Bill by Day",
                  xaxis_title="Day",
                  yaxis_title="Total Bill")

# Display the plot
pio.show(fig)
```

```python
## Answer 3)
import plotly.express as px

# Load the "tips" dataset from Plotly
tips = px.data.tips()

# Plot histogram using Plotly Express
fig = px.histogram(tips, x="sex", y="total_bill", color="day", pattern_shape="smoker",
                   labels={"sex": "Sex", "total_bill": "Total Bill"})

# Update plot layout and display the figure
fig.update_layout(title="Histogram of Total Bill by Sex",
                  xaxis_title="Sex",
                  yaxis_title="Total Bill")

# Display the plot
fig.show()
```

```python
## Answer 4)

import plotly.express as px

# Load the "iris" dataset from Plotly
iris = px.data.iris()

# Plot scatter matrix using Plotly Express
fig = px.scatter_matrix(iris, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
                        color="species", title="Scatter Matrix of Iris Dataset")

# Display the plot
fig.show()
```

```python
## Answer 5)

import plotly.express as px

# Load sample data
data = [1, 1, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5, 6, 6, 7, 8, 9, 9, 10]

# Plot distplot using Plotly Express
fig = px.histogram(data, marginal="rug", title="Distribution Plot")
fig.update_layout(xaxis_title="Value", yaxis_title="Count")

# Display the plot
fig.show()
```
