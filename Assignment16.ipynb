```python
#Answer 1)
MongoDB is a popular NoSQL document-oriented database that stores data in flexible, JSON-like documents with dynamic schemas. Instead of using tables and rows as in traditional SQL databases, MongoDB stores data in collections and documents, which can contain fields of any data type including arrays, binary data, and sub-documents.

Non-relational databases, also known as NoSQL databases, are a type of database that does not use the traditional tabular structure of relational databases. Instead, they use a variety of data models, such as key-value, document, column-family, or graph. Non-relational databases are designed to handle large amounts of unstructured and semi-structured data, making them suitable for use cases such as real-time analytics, content management, and social media.

MongoDB is often preferred over SQL databases in the following scenarios:

Handling unstructured data: MongoDB is a good choice for handling unstructured and semi-structured data, such as social media data or IoT data, where the structure of the data may change frequently.

Horizontal scalability: MongoDB's distributed architecture allows for easy horizontal scalability, which makes it well-suited for large, high-traffic applications.

Developer productivity: MongoDB's flexible document-based data model allows for faster development and iteration, as developers can work with data in a format that more closely resembles their code.

Cloud-native applications: MongoDB is often used in modern cloud-native applications because of its compatibility with cloud platforms like AWS, Azure, and Google Cloud Platform.

Real-time analytics: MongoDB's aggregation framework allows for real-time analytics on large datasets, making it a popular choice for applications that require real-time insights.

In summary, MongoDB is a NoSQL document-oriented database that is well-suited for handling unstructured and semi-structured data, horizontal scalability, and developer productivity, making it a popular choice for modern, cloud-native applications.
```

```python
#Answer 6)
The sort() method in MongoDB is used to sort the results of a query in either ascending or descending order. 
import pymongo

# Connect to MongoDB database
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["mydatabase"]
collection = db["mycollection"]

# Define a query to find all documents in the collection and sort by age in descending order
query = {}
sort = {"age": -1}

# Use the find() method to execute the query and retrieve the documents, sorted by age
results = collection.find(query).sort(sort)

# Print out the results
for result in results:
    print(result)
```

```python
# Answer 7)
In PyMongo, the delete_one(), delete_many(), and drop() methods are used to remove documents or entire collections from a MongoDB database. Here's a brief explanation of each method:

delete_one(filter, collation=None) - This method removes the first document that matches the specified filter from the collection. If there are multiple documents that match the filter, only the first one will be deleted. The filter parameter is a dictionary that specifies the criteria for the documents to be deleted, and the collation parameter is an optional parameter that specifies the collation rules for string comparison.

delete_many(filter, collation=None) - This method removes all documents that match the specified filter from the collection. The filter and collation parameters have the same meaning as for delete_one().

drop() - This method removes the entire collection from the database. The collection will be deleted permanently, along with all of its documents and indexes.
```
