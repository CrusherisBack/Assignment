```python
# Answer 3) In Flask, app routing refers to the process of mapping a URL or route to a view function that handles the request and returns a response to the client.

When a user requests a URL in a Flask application, the Flask web server uses the URL to determine which view function should handle the request. The view function then generates a response, which is returned to the user as an HTTP response.

We use app routes in Flask to define the URLs or endpoints that the application will respond to. App routes are typically defined using the @app.route decorator, which maps a URL pattern to a Python function.
```

# Answer 5)
In Flask, the url_for() function is used for URL building. This function generates a URL based on the name of a view function and the arguments provided to it. The url_for() function takes the name of a view function as its first argument, followed by keyword arguments representing any variable parts of the URL.

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/user/<username>')
def profile(username):
    return f'Hello, {username}!'

with app.test_request_context():
    print(url_for('profile', username='Alice'))
    print(url_for('profile', username='Bob'))
