Seaborn is a popular data visualization library in Python that is built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Here are five commonly used plots in Seaborn along with their uses:

1. Scatter Plot:
   - Uses: Scatter plots are used to visualize the relationship between two continuous variables. They help identify patterns, trends, and outliers in the data. They are also useful for exploring correlations and assessing the strength of relationships between variables.

2. Histogram:
   - Uses: Histograms display the distribution of a single variable by dividing the data into bins and showing the frequency or count of values falling into each bin. They are helpful in understanding the shape, central tendency, and spread of the data.

3. Bar Plot:
   - Uses: Bar plots are used to compare categorical variables and display their counts or aggregated values. They are effective in visualizing and comparing the distribution of categories or groups. Bar plots can be customized to show means, medians, or any other statistical measures.

4. Box Plot:
   - Uses: Box plots, also known as box-and-whisker plots, provide a summary of the distribution of a continuous variable. They display the median, quartiles, and any outliers in the data. Box plots are useful for identifying skewness, variability, and comparing distributions between different categories.

5. Heatmap:
   - Uses: Heatmaps are graphical representations of data in a matrix form, where values are represented by colors. They are often used to visualize correlations between variables in a dataset. Heatmaps are also useful for showing patterns and relationships in large datasets.



```python
## Answer 2)
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri = sns.load_dataset("fmri")

# Plot line plot using x = "timepoint" and y = "signal" for different events and regions
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="event", style="region")

# Set plot title and axes labels
plt.title("FMRI Signal over Time")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

# Display the plot
plt.show()
```

```python
## Answer 3)

import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic = sns.load_dataset("titanic")

# Plot box plots using x = 'pclass', y = 'age'
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title("Box Plot of Age by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")

# Plot box plots using x = 'pclass', y = 'fare'
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title("Box Plot of Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")

# Adjust the spacing between subplots
plt.tight_layout()

# Display the plots
plt.show()
```

```python
## Answer 4)

import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds = sns.load_dataset("diamonds")

# Plot histogram with hue for the 'cut' column
sns.histplot(data=diamonds, x='price', hue='cut', kde=True)

# Set plot title and axes labels
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Count")

# Display the plot
plt.show()

```

```python
## Answer 5)

import seaborn as sns

# Load the "iris" dataset
iris = sns.load_dataset("iris")

# Plot pair plot with hue for the 'species' column
sns.pairplot(data=iris, hue='species')

# Display the plot
plt.show()
```

```python
## Answer 6)

import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights = sns.load_dataset("flights")

# Reshape the dataset for heatmap
flights_pivot = flights.pivot("month", "year", "passengers")

# Plot the heatmap
sns.heatmap(flights_pivot, cmap="YlGnBu")

# Set plot title and axes labels
plt.title("Flights Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")

# Display the plot
plt.show()
```
